```python
"""
NeuroScan AI Version Alpha
Created By Ishan Leung

Built Using Tensorflow and Python 3

Instructions:
(1) Drag keras_model.h5, labels.txt, and testing data (image files, 100 max)
to files tab. Note: Drag files directly and DO NOT put the files into a
subfolder and drag that subfolder over. Ensure the First File is named
image(1).jpg and the Last File is named image(max).jpg
Note: Replace 'max' with the highest file number.

(2) Click Runtime --> Run All

(3) Enter Number of Files

(4) Select Tumor Type to Test For

(5) Interpret Results
"""

from keras.models import load_model  # TensorFlow is required for Keras to work
from PIL import Image, ImageOps  # Install pillow instead of PIL
import numpy as np

num = input("How Many Files (1-100):")
files = []
type_tested = int(input("What Type are you testing?\n0 for No Tumor\n1 for Glioma Tumor\n2 for Meningioma Tumor\n3 for Pituitary Tumor"))
if type_tested == 0:
  type_tested = "No Tumor"
elif type_tested == 1:
  type_tested = "Glioma Tumor"
elif type_tested == 2:
  type_tested = "Meningioma Tumor"
elif type_tested == 3:
  type_tested = "Pituitary Tumor"
# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

#Add Lines to Increase Legibility
print()
print()
print("Running Model...")
print()

# Load the model
model = load_model("/content/keras_model.h5", compile=False)

# Load the labels
class_names = open("/content/labels.txt", "r").readlines()

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

print("Testing For: " + type_tested)
print()

for i in range (1,int(num)+1):
  # Replace this with the path to your image
  image = Image.open("/content/image("+ num +").jpg").convert("RGB")

  # resizing the image to be at least 224x224 and then cropping from the center
  size = (224, 224)
  image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)

  # turn the image into a numpy array
  image_array = np.asarray(image)

  # Normalize the image
  normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1

  # Load the image into the array
  data[0] = normalized_image_array

  # Predicts the model
  prediction = model.predict(data)
  index = np.argmax(prediction)
  class_name = class_names[index]
  confidence_score = prediction[0][index]

  # Print prediction and confidence score
  print("File Number " + str(i))
  print("Class:", class_name[2:], end="")
  print("Confidence Score:", confidence_score)
```
